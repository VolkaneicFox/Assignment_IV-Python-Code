```python
import pandas as pd

df_10 = pd.read_excel('politics_for_the_web.xls',sheet_name = 'data for table 10')
df_11 = pd.read_excel('politics_for_the_web.xls',sheet_name = 'data for table 11')
df_10['constant'] = 1
df_11['constant'] =1
df_10.head()
```

```python
endogenous = ['Executive constraints (1960-2000)',
       'Expropriation risk (1982-1997)', 'Autocracy -- Alvarez (1960-1990)',
       'Government effectiveness (1998-2000)']
exogenous = ['Log settler mortality',
       'Population at risk of malaria (1994)']
```

```python
import statsmodels.api as sm
```

```python
for var_en in endogenous:
    df = df_10[[var_en,'constant']+exogenous].dropna()
    model = sm.OLS(df[var_en],df[exogenous+['constant']]).fit()
    print (model.summary())
```

```python
exogenous = [['Share of population living in temperate zone (1995)',
       'French legal origin', 'Log population density in 1500',
              'constant'],['Share of population living in temperate zone (1995)',
       'French legal origin', 'Log settler mortality', 'constant'],
            ['Years of schooling (1960-2000)', 'Executive constraints (1960-2000)',
       'Share of population living in temperate zone (1995)', 'constant']]
endogenous = [[ 'Years of schooling (1960-2000)',
       'Executive constraints (1960-2000)'], ' Log GDP per capita in 2000']
```

```python
#first stage regresses endogenous[0] on exogenous[0]
for endo in endogenous[0]:
    df = df_11.dropna()
    model = sm.OLS(df[endo],df[exogenous[0]]).fit()
    print(model.summary())
    df = df_11[[endo]+exogenous[1]].dropna()
    model = sm.OLS(df[endo],df[exogenous[1]]).fit()
    print(model.summary())
```

```python
df = df_11.dropna()
```

```python
df.columns
```

The endogenous variable is 'Log GDP per capita in 2000' The regressors are 'Years of schooling (1960-2000)','Executive constraints (1960-2000)' and 'Share of population living in temperate zone (1995)'. The endogeneity problem arises in 'Executive constraints (1960-2000)' and 'Years of schooling (1960-2000). The first stage regresses the endogenous regressors on the instruments, 'Share of population living in temperate zone (1995)', 'French legal origin', 'Log population density in 1500' and 'Log settler mortality'.

```python
first_stage_models = []
```

```python
model = sm.OLS(df['Executive constraints (1960-2000)'],df[['Share of population living in temperate zone (1995)',
                                                           'French legal origin', 'Log settler mortality','constant']])
model = model.fit()
first_stage_models.append(model)
```

```python
print(first_stage_models[0].summary())
```

```python
df['pred Executive Constraints (1960-2000)'] = first_stage_models[0].predict()
```

```python
model = sm.OLS(df['Years of schooling (1960-2000)'],df[['Share of population living in temperate zone (1995)',
                                                           'French legal origin', 'Log settler mortality','constant']])
model = model.fit()
first_stage_models.append(model)
```

```python
print(first_stage_models[1].summary())
```

```python
df['pred Years of schooling (1960-2000)'] = first_stage_models[1].predict()
```

```python
modelFinal = sm.OLS(df[' Log GDP per capita in 2000'], df[['constant','pred Executive Constraints (1960-2000)',
                                                           'pred Years of schooling (1960-2000)',
                                                           'Share of population living in temperate zone (1995)']]).fit()
```

```python
df = df_11[['constant','Executive constraints (1960-2000)', 'Years of schooling (1960-2000)', 
            ' Log GDP per capita in 2000','Share of population living in temperate zone (1995)',
            'French legal origin','Log population density in 1500']]
```

```python
df.dropna(inplace = True)
```

```python
df.head()
```

```python
instruments = ['constant','Share of population living in temperate zone (1995)','French legal origin',
               'Log population density in 1500']
endo_1 = ['Executive constraints (1960-2000)', 'Years of schooling (1960-2000)']
```

```python
first_stage_models = []
for var_en in endo_1:
    model = sm.OLS(df[var_en],df[instruments]).fit()
    first_stage_models.append(model)
    print (model.summary())
```

```python
df['pred Executive constraints (1960-2000)'] = first_stage_models[0].predict()
df['pred Years of schooling (1960-2000)'] = first_stage_models[1].predict()
```

```python
df.head()
```

```python
model_final = sm.OLS(df[' Log GDP per capita in 2000'], df[['Share of population living in temperate zone (1995)',
                                                            'pred Executive constraints (1960-2000)', 
                                                            'pred Years of schooling (1960-2000)','constant']])
```

```python
model_final = model_final.fit()
print (model_final.summary())
```
